```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Functions
def logit(p):
    assert p <= 1.0
    assert p >= 0.0
    return np.log2(p / (1 - p))

def log(p):
    assert p >= 0
    return np.log2(p)

def sigmoid(p):
    return (1 / (1 + np.exp(-p)))
```

```python
#Set size of figure
plt.figure(figsize=(20,10))

#Logit
plt.subplot(221)
X = np.arange(0, 1, 0.01)
Y = map(logit, X)
plt.plot(X, Y)
plt.title("Logit", fontdict={'fontsize':20})
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

#Log
plt.subplot(222)
X = np.arange(0, 5, 0.05)
Y = map(log, X)
plt.plot(X, Y)
plt.title("Log", fontdict={'fontsize':20})
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

#Sigmoid
plt.subplot(223)
X = np.arange(-5, 5, 0.05)
Y = map(sigmoid, X)
plt.plot(X, Y)
plt.title("Sigmoid", fontdict={'fontsize':20})
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

#Set properties

plt.subplots_adjust(top=0.90, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()
```
